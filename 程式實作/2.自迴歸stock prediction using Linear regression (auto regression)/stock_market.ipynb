## Step 1 得到資料: Get Data

```python
#import套件
import pandas as pd
import matplotlib.pyplot as plt

#我們的開發環境是 Jupyter Notebook，這個指令可以讓圖形不會在新視窗呈現。
%matplotlib inline

#讀入csv文字檔
data=pd.read_csv('2330.csv')

print(data.head(10))

#取我們要的欄位
data2 = data[['Date','close']]
data2.head()
#利用位移來準備(明天)訓練資料
data2['x1']=data2['close'].shift(1)
data2['x2']=data2['close'].shift(2)
data2['x3']=data2['close'].shift(3)
data2['x4']=data2['close'].shift(4)
data2['x5']=data2['close'].shift(5)

#去除nan
data2=data2.dropna()
data.head()
```

## Step 2 處理資料: Prepare X, Y

```python
#只擷取需要欄位
x=data2[['x1','x2','x3','x4','x5']].values.reshape(-1,5)
y=data2['close'].values.reshape(-1,1)
#print(x.shape)#(4946, 5)
```

## Step 3 建立線性迴歸分析模型:  Model training

```python
#Python 機器學習套件 scikit-learn
#使用 sklearn.linear_model 的 LinearRegression() 線性迴歸方法
from sklearn.linear_model import LinearRegression as LR  # 線性迴歸
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

#模型初始化
model=LR()

#訓練用fit
model.fit(x,y)

'''
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import  r2_score as R2
import numpy as np
max =0
for _C in np.arange(0.1,1,0.1):
    for _degree  in range(1,10):
        model = SVR(kernel='linear',C=_C,degree =_degree)
        model.fit(X,Y)
        preY=model.predict(X)
        if R2(preY,Y)> max:
            max =R2(preY,Y)
            print('r2=',R2(preY,Y),'C=',_C,'degree=',_degree) 
'''
```

## Step 4 評估模型: Evaluate Model

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import  r2_score as R2

#透過.score方法返回判定係數評估模型好不好(1較好)
score=model.score(x,y)

print('score:',score)#0.9882683991934083
print('r2=',R2(texty,y)) 
print('mse=',MSE(texty,y))
```
